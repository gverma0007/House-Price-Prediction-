```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"train.csv")
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
rem_col=df.isnull().sum()/df.shape[0]*100
```

```python
null_coll_lst= rem_col[rem_col>20].index
null_coll_lst
```

```python
df = df.drop(columns= null_coll_lst)
```

```python
df.shape
```

```python

df.head()
```

```python
df_num = df.select_dtypes(include = ['int', 'float'])
```

```python
df_num.head()
```

```python
n_col = df_num.columns
n_col
```

```python
df_num1=df_num.copy()
for i in n_col:
    df_num1[i]= df_num1[i].replace(np.nan,df_num1[i].median())
```

```python
df_num1.isnull().sum().sum()
```

```python
df_cat= df.select_dtypes(include=['object'])
```

```python
df_cat.isnull().sum().sum()
```

```python
df_cat1=df_cat.copy()
for i in df_cat1.columns:
    df_cat1[i]= df_cat1[i].replace(np.nan , df_cat1[i].mode()[0])
```

```python
df_cat1.isnull().sum().sum()
```

```python
df_cat1.columns
```

```python
for i in df_cat1.columns:
    print(df_cat1[i].value_counts())
    print('---------------------------------------------------')
```

```python
q_map = {
    'Po':1,
    'Fa':2,
    'TA':3,
    'Gd':4,
    'Ex':5,
    'No':1,
    'Mn':2,
    'Av':3,   
}
```

```python
col = ['ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','HeatingQC','KitchenQual','GarageQual','GarageCond']
for i in col:
    l = df_cat1[i].map(q_map)
    df_num1 = pd.concat([df_num1 , pd.DataFrame({i+'_en':l})],axis =1)
```

```python
# df_cat1 = df_cat1.drop(columns=col)
display(df_cat1.columns , df_num1.isnull().sum())
```

```python
for i in df_cat1.columns:
    print(df_cat1[i].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in df_cat1.columns:
    if(len(df_cat1[i].unique())<=4):
        arr = le.fit_transform(df_cat1[i])
        df_num1[i+'_len'] = arr
    else:
        a = df_cat1[i].value_counts()
        df_num1[i+'_freq']=df_cat1[i].map(a)
```

```python
df_num1.isnull().sum().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
mms = MinMaxScaler()
```

```python
X= df_num1.drop(columns= ['SalePrice'])
y= df_num1['SalePrice']
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
df_num1.isnull().sum().sum()
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
X_train_sc = pd.DataFrame(ss.fit_transform(X_train),columns= X_train.columns)
X_test_sc = pd.DataFrame(ss.fit_transform(X_test),columns= X_test.columns)
```

```python
X_test_sc
```

```python
lr.fit(X_train_sc,y_train)
```

```python
y_pred=lr.predict(X_test_sc)
```

```python
lr.score(X_test_sc,y_test)
```

```python
su = 0
for i in range (20):
    X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
    X_train_sc = pd.DataFrame(ss.fit_transform(X_train),columns= X_train.columns)
    X_test_sc = pd.DataFrame(ss.fit_transform(X_test),columns= X_test.columns)
    lr.fit(X_train_sc,y_train)
    h = lr.score(X_test_sc,y_test)
    su = su +h
    print(h*100)

print('avg==',su/20*100)    
```
